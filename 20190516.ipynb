```python
math = [
    {"name":"Jeff", "score":12},
    {"name":"Leo", "score":63},
    {"name":"Kevin", "score":100},
    {"name":"Apple", "score":77},
    {"name":"Elle", "score":84},
    {"name":"Emma", "score":60},
]

eng = [
    {"name":"Jeff", "score":58},
    {"name":"Leo", "score":100},
    {"name":"Kevin", "score":97},
    {"name":"Apple", "score":85},
    {"name":"Elle", "score":92},
    {"name":"Emma", "score":41},
]

def getScoresSum(data):
    result = 0
    
    for ele in data:
        result += ele["score"]
    return result

def getScoresAve(data):
    result = 0 
    for ele in data:
        result += ele["score"]
    return result / len (data)


def getFailedList(data):
    result = []
    for ele in data:
        if ele["score"] < 60:
            result.append(ele["name"])
    return

mathSum= getScoresSum(math)
engAve = getScoresAve(eng)
mathFailed = getFailedList(math)

print(mathSum)
print(engAve)
print(mathFailed)
```

```python
#類別

class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def sayMyName(self):
        print("HiHi, I am {}".format(self.name))
    def showMyAge(self):
        print("I am {} years old.".format(self.age))
        
person1 = Person("Jeff",18)
person2 = Person("Leo",22)


person1.sayMyName()
person1.showMyAge()
```

```python
class Shoe:
    city = "Taipei"
    
    #先定義class的動態物件內容，__init__是用來定義物件，第一個參數是物件（必為self)，接下來就是屬性（如size,price）
    def __init__(self,size,price):
        self.size = size
        self.price = price
    
    def showMySize(self):
            print("My size is",format(self.size))
    
    def showThePrice(self):
            print("This Shoe costs",format(self.price))

shoe1 = Shoe(9,1000)
shoe2 = Shoe(12,2500)

shoe1.showMySize()
shoe2.showThePrice()
```

```python
### By Class
class EngClass:
    def __init__(self,data):
        self.data = data
        
    def getScoresAvg(self):
        result = 0
        for ele in self.data:
            result += ele["score"]
        return result / len(self.data)
    
    
class MathClass:
    def __init__(self,data):
        self.data = data
        
    def getScoresAvg(self):
        result = 0
        for ele in self.data:
            result += ele["score"]
        return result / len(self.data)
        

mathStudScores = [
    {"name":"Jeff", "score":95},    
    {"name":"Leo" , "score":63},
    {"name":"Keven", "score":100},
    {"name":"Apple", "score":77},
    {"name":"Elle", "score":84},
    {"name":"Emma", "score":60},
]

engStudScores = [
    {"name":"Jeff","score":58},    
    {"name":"Leo", "score":100},
    {"name":"Keven","score":97},
    {"name":"Apple","score":85},
    {"name":"Elle","score":92},
    {"name":"Emma","score":41},
]

# 數學 : 總合
# 英文 : 平均 & 不及格名單

###
# Definition
### Mathsult
    
###
# Main
engClass = EngClass(engStudScores)
print("Eng Avg : ",engClass.getScoresAvg())

mathClass = MathClass(mathStudScores)
print("Math Avg : ",mathClass.getScoresAvg())
```
